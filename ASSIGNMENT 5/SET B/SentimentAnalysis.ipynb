```python
df = data.drop('Rating', axis=1)
```

```python
df = data.drop('Manufacturer Response', axis=1)
df
```

```python
Category_mode = df['Category'].mode()
Category_mode
```

```python
Sub_Category_mode = df['Sub Category'].mode()
Sub_Category_mode
```

```python
Product_Description_mode = df['Product Description'].mode()
Product_Description_mode
```

```python
df['Category'].fillna(Category_mode,inplace=True)
df['Sub Category'].fillna(Sub_Category_mode,inplace=True)
df['Product Description'].fillna(Product_Description_mode,inplace=True)
```
